```python
import numpy as np


ldpc_design_file = open('commpy/channelcoding/designs/ldpc/gallager/96.33.964.txt')

[n_vnodes, n_cnodes] = [int(x) for x in ldpc_design_file.readline().split(' ')]
[max_vnode_deg, max_cnode_deg] = [int(x) for x in ldpc_design_file.readline().split(' ')]
vnode_deg_list = np.array([int(x) for x in ldpc_design_file.readline().split(' ')[:-1]], np.int32)
cnode_deg_list = np.array([int(x) for x in ldpc_design_file.readline().split(' ')[:-1]], np.int32)

cnode_adj_list = -np.ones([n_cnodes, max_cnode_deg], int)
vnode_adj_list = -np.ones([n_vnodes, max_vnode_deg], int)

for vnode_idx in xrange(n_vnodes):
    vnode_adj_list[vnode_idx, 0:vnode_deg_list[vnode_idx]] = \
        np.array([int(x)-1 for x in ldpc_design_file.readline().split('\t')])

for cnode_idx in xrange(n_cnodes):
    cnode_adj_list[cnode_idx, 0:cnode_deg_list[cnode_idx]] = \
        np.array([int(x)-1 for x in ldpc_design_file.readline().split('\t')])
        
cnode_vnode_map = -np.ones([n_cnodes, max_cnode_deg], int)
vnode_cnode_map = -np.ones([n_vnodes, max_vnode_deg], int)
cnode_list = np.arange(n_cnodes)
vnode_list = np.arange(n_vnodes)

for cnode in xrange(n_cnodes):
    for i, vnode in enumerate(cnode_adj_list[cnode, 0:cnode_deg_list[cnode]]):
        cnode_vnode_map[cnode, i] = cnode_list[vnode_adj_list[vnode, :] == cnode]

for vnode in xrange(n_vnodes):
    for i, cnode in enumerate(vnode_adj_list[vnode, 0:vnode_deg_list[vnode]]):
        vnode_cnode_map[vnode, i] = vnode_list[cnode_adj_list[cnode, :] == vnode]

        
cnode_adj_list_1d = cnode_adj_list.flatten().astype(np.int32)
vnode_adj_list_1d = vnode_adj_list.flatten().astype(np.int32)
cnode_vnode_map_1d = cnode_vnode_map.flatten().astype(np.int32)
vnode_cnode_map_1d = vnode_cnode_map.flatten().astype(np.int32)

pmat = np.zeros([n_cnodes, n_vnodes], int)
for cnode_idx in xrange(n_cnodes):
    pmat[cnode_idx, cnode_adj_list[cnode_idx, :]] = 1

#print n_cnodes, n_vnodes
#print max_cnode_deg, max_vnode_deg
#print cnode_adj_list
#print vnode_adj_list
#print cnode_vnode_map
#print vnode_cnode_map
```

```python
import numpy as np 
from commpy.channelcoding.ldpc import ldpc_bp_decode, get_ldpc_code_params
from commpy.utilities import hamming_dist, bitarray2dec, dec2bitarray

ldpc_design_filename = 'commpy/channelcoding/designs/ldpc/gallager/96.33.964.txt'
ldpc_code_params = get_ldpc_code_params(ldpc_design_filename)

N = 96
k = 48

rate = 0.5
Es = 1.0

snr_list = np.array([1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0])
niters = 1000000

tx_codeword = np.zeros(N, int)
ldpcbp_iters = 100
dec_algo_type = 'MSA'
#cost_function = np.zeros(N)

for ebno in snr_list[0:5]:

    noise_std = 1/np.sqrt((10**(ebno/10.0))*rate*2/Es)    
    fer_cnt_bp = 0
    fer_cnt_lp = 0
    
    for iter_cnt in xrange(niters):
        
        awgn_array = noise_std * np.random.randn(N)
        rx_word = 1-(2*tx_codeword) + awgn_array 
        rx_llrs = 2.0*rx_word/(noise_std**2)
        
        [dec_word, out_llrs] = ldpc_bp_decode(rx_llrs, ldpc_code_params, dec_algo_type, ldpcbp_iters)
        
        num_bit_errors = hamming_dist(tx_codeword, dec_word)
        if num_bit_errors > 0:
            fer_cnt_bp += 1
                
        if (iter_cnt+1) % 1000 == 0:
            print fer_cnt_bp, iter_cnt+1
            
        if fer_cnt_bp >= 200:
            break

    print "BAWGNC(", ebno, ") -- ", fer_cnt_bp, iter_cnt+1 
    print
```

```python
import sys, time
from numpy import array, arange, power, log10, zeros, sqrt
from numpy.random import randint, randn
from commpy.channelcoding import turbo_encode, map_decode, Trellis, turbo_decode
from commpy.utilities import hamming_dist
import commpy.channelcoding.interleavers as ii


total_bits = 1000000
block_size = 10000
num_blocks = total_bits/block_size

# Noise Variance
noise_var_array = float(3)/(2*power(10, arange(0, 0.55, 0.05)))

noise_var = noise_var_array[5:6]

# Compute Eb/No using noise variance and Eb = 3 (for BPSK)
Eb_No_dB = 10 * log10(float(3)/(2*noise_var))

prob_bit_error = zeros(len(Eb_No_dB))

number_iterations = 4

# Encoder Definition
memory = array([2])
g_matrix = array([[04, 07]])
feedback = 05
trellis = Trellis(memory, g_matrix, feedback, 'rsc')
interlv_seed = 8368 

interleaver = ii.RandInterlv(block_size, interlv_seed)

print " Turbo Decoder "

L_int = zeros(block_size)

for i in xrange(len(noise_var)):
    
    total_bit_errors = 0

    print "=================================="
    print "SNR: " + str(Eb_No_dB[i])

    print "======== Iteration " + str(i) + " =========" 
    print "Progress: ",

    for block_num in xrange(num_blocks):

        msg_bits = randint(0, 2, block_size)
        
        [sys_stream, non_sys_stream_1, non_sys_stream_2] = \
                turbo_encode(msg_bits, trellis, trellis, interleaver)
    
        #sys_stream = sys_stream[0:-2]
        #non_sys_stream_1 = non_sys_stream_1[0:-2]
        #non_sys_stream_2 = non_sys_stream_2[0:-2]
    
        #print len(sys_stream), len(non_sys_stream_1), len(non_sys_stream_2)
        awgn = sqrt(noise_var[i]) * randn((block_size+2)*3)
        
        rx_sys_stream = (2*sys_stream - 1) + awgn[0:block_size]
        rx_non_sys_stream_1 = (2*non_sys_stream_1 - 1) + \
                awgn[block_size: (block_size)*2]
        rx_non_sys_stream_2 = (2*non_sys_stream_2 - 1) + \
                awgn[(block_size)*2 : (block_size)*3]
            
        L_int[:] = 0.0
        decoded_bits = turbo_decode(rx_sys_stream, rx_non_sys_stream_1, 
                                    rx_non_sys_stream_2, trellis, 
                                    noise_var[i], number_iterations, 
                                    interleaver, L_int)

        block_bit_errors = hamming_dist(sys_stream, decoded_bits)
        total_bit_errors += block_bit_errors
        
        if (block_num+1) % 20 == 0:
            sys.stdout.write('.') 

    prob_bit_error[i] = float(total_bit_errors)/((block_size)*num_blocks)
    
    print
    print "Pe: " + str(prob_bit_error[i])

```

```python
import sys, time
from numpy import array, arange, power, log10, zeros, sqrt
from numpy.random import randint, randn
from commpy.channelcoding import turbo_encode, map_decode, Trellis, turbo_decode
from commpy.utilities import hamming_dist
import commpy.channelcoding.interleavers as ii

block_size = 10000
number_iterations = 4
noise_var = 0.2

# Encoder Definition
memory = array([2])
g_matrix = array([[04, 07]])
feedback = 05
trellis = Trellis(memory, g_matrix, feedback, 'rsc')
interlv_seed = 8368 

interleaver = ii.RandInterlv(block_size, interlv_seed)

msg_bits = randint(0, 2, block_size)
        
[sys_stream, non_sys_stream_1, non_sys_stream_2] = turbo_encode(msg_bits, trellis, trellis, interleaver)

awgn = sqrt(noise_var) * randn((block_size+2)*3)
        
rx_sys_stream = (2*sys_stream - 1) + awgn[0:block_size]
rx_non_sys_stream_1 = (2*non_sys_stream_1 - 1) + awgn[block_size: (block_size)*2]
rx_non_sys_stream_2 = (2*non_sys_stream_2 - 1) + awgn[(block_size)*2 : (block_size)*3]
          
L_int = zeros(block_size)
L_int[:] = 0.0
decoded_bits = turbo_decode(rx_sys_stream, rx_non_sys_stream_1, rx_non_sys_stream_2, trellis, 
                            noise_var, number_iterations, interleaver, L_int)

block_bit_errors = hamming_dist(sys_stream, decoded_bits)

print "Number of Errors (after decoding): ", block_bit_errors
```
